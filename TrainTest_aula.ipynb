Dados fictícios para teste de treino e teste.

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(2)

pageSpeed = np.random.normal(3.0, 1.0, 100)
purchaseAmount = np.random.normal(50.0, 30.0, 100) / pageSpeed
plt.scatter(pageSpeed, purchaseAmount)
```

Separação de dados dentro do array para dizer quem vai ser de teste e quem vai ser de treinamento

```python
trainX = pageSpeed[:80]
testX = pageSpeed[80:]

trainY = purchaseAmount[:80]
testY = purchaseAmount[80:]
```

Amostra de treinamento

```python
plt.scatter(trainX, trainY)
```

Amostra de teste

```python
plt.scatter(testX, testY)
```

Criação de um polinomio de 8 grau, com o intuito de causar um overfitting

```python
x = np.array(trainX)
y = np.array(trainY)

p4 = np.poly1d(np.polyfit(x,y,8))
```

```python
xp =  np.linspace(0,7,100)
axes = plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([0,200])
plt.scatter(x,y)
plt.plot(xp, p4(xp), c='r')
plt.show()
```

Sem o overfitting mas ainda longe de ser perfeito

```python
testx =  np.array(testX)
testy = np.array(testY)
axes = plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([0,200])
plt.scatter(testx, testy)
plt.plot(xp, p4(xp), c='r')
plt.show()
```

R-quadrado de 0.3, sendo bem ruim para poder prever futuros dados

```python
from sklearn.metrics import r2_score

r2 = r2_score(testy , p4(testx))
print(r2)
```

R-quadrado razoável conseguiria prever dados novos mas com alguns erros

```python
r2 = r2_score(np.array(trainY) , p4(np.array(trainX)))
print(r2)
```
