Criação de dados fictícios para o teste da regressão linear, ele relaciona a velocidade de carregar uma página com o tanto de dinheiro gastos.

```python
import matplotlib.pyplot as plt
import numpy as np
pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = 100 - (pageSpeeds + np.random.normal(0, 0.1, 1000)) * 3
plt.scatter(pageSpeeds, purchaseAmount)
```

Importando a biblioteca para poder executar as funções.

```python
from scipy import stats 
slope, intercept, r_value, p_value, std_err = stats.linregress(pageSpeeds, purchaseAmount)
```

R-Quadrado mostrando que a regressão linear esta sendo bem aplicada.

```python
r_value ** 2
```

Função para criar a linha da regresão.

```python
def predict(x):
    return slope * x + intercept

fitline = predict(pageSpeeds)
plt.scatter(pageSpeeds, purchaseAmount)
plt.plot(pageSpeeds, fitline, c ='r')
plt.show()
```
